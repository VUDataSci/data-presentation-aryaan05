```python
#reading the dataset in
df = pd.read_csv("bachelorette.csv")
```

```python
elimination_columns = [col for col in df.columns if 'ELIMINATION' in col]
```

```python



for col in elimination_columns:
    # 'E' codes indicate elimination (1)
    # 'R' codes and NaN indicate survival (0)
    df[col] = df[col].apply(lambda x: 1 if isinstance(x, str) and x.startswith('E') else 0)
```

```python
df.head(15)
```

```python
# Identify columns with elimination information
elimination_columns = [col for col in df.columns if 'ELIMINATION' in col]

# Update DataFrame to reflect eliminations and roses according to new rules
for col in elimination_columns:
    df[col + '_Eliminated'] = df[col].apply(lambda x: 1 if isinstance(x, str) and x[0] == 'E' else 0)
    df[col + '_Rose'] = df[col].apply(lambda x: 1 if isinstance(x, str) and x[0] == 'R' else 0)

# Calculate survival duration
df['Survival Duration'] = df[elimination_columns].apply(lambda row: len(elimination_columns) - sum(row), axis=1)

```

```python
# Compute descriptive statistics for the updated survival duration
survival_stats_updated = df['Survival Duration'].describe()

survival_stats_updated
```

```python
# Aggregate the number of dates per contestant
date_columns = [col for col in df.columns if 'DATES' in col]
df['Total Dates'] = df[date_columns].apply(lambda row: sum(row.notna()), axis=1)

# Correlation between 'Total Dates' and 'Survival Duration'
correlation_stats_updated = df[['Total Dates', 'Survival Duration']].corr()
```

```python
# Plot the updated correlation with a scatter plot and regression line
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
scatter_plot_updated = sns.regplot(x='Total Dates', y='Survival Duration', data=df,
                                   scatter_kws={'alpha':0.6}, line_kws={'color': 'red'})
plt.title('Updated Correlation Between Number of Dates and Survival Duration')
plt.xlabel('Total Number of Dates')
plt.ylabel('Survival Duration (Rounds)')
plt.show()
```

```python
import networkx as nx
from itertools import combinations
import matplotlib.pyplot as plt

# Assuming df is the DataFrame with the correct bachelorette data loaded
# We will also assume that the 'DATES' columns contain information about group dates
date_columns = [col for col in df.columns if col.startswith('DATES')]

# Create a graph from the pairs of contestants on the same dates
G_updated = nx.Graph()

# Add edges based on group dates
for col in date_columns:
    # Here we consider only the entries that have a 'D' indicating a group date
    contestants_on_date = df[df[col].str.startswith('D', na=False)]['CONTESTANT'].tolist()
    # Create all possible pairs from the contestants on the date
    pairs_on_date = list(combinations(contestants_on_date, 2))
    # Add these pairs as edges to the graph
    G_updated.add_edges_from(pairs_on_date)

# Draw the updated network graph without labels
plt.figure(figsize=(12, 12))
pos_updated = nx.spring_layout(G_updated, seed=7)
nx.draw(G_updated, pos=pos_updated, node_color='blue', edge_color='gray', alpha=0.6)
plt.title('Updated Network of Contestant Connections on Group Dates')
plt.axis('off')
plt.show()


```
